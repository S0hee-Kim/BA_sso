## 2. 분석 개요
* 네트워크 : 각각의 객체들이 상호 연결된 구조
* 구성요소 : node(Node, vertex)와 링크(link, edge)
* 노드 : 고유한 속성을 가지고 있는 행위자
* 링크 : 노드들간의 관계(relation)
> 네트워크는 복수의 노드들이 연결된 구조

* 예시) 인터넷 망, 인터넷 웹, 소셜미디어, 친구들간의 관계, 지하철 노선도, 지도 상의 도로망, 항공 노선도, 가계도(족보), 장바구니 분석, 단톡방 대화 분석

    * 직업을 얻게 된 대부분의 사람들은 친구나 친척처럼 가까운 사람들을 통해서가 아니라 단순히 그냥 가볍게 알고 지내는 지인을 통해서 얻은 경우가 훨씬 더 많았음

* **프리만의 '중심성 지표' 개발**

* **버트의 '구조적 공백'**
    * 사회 네트워크 구조 내에서 상대적으로 중요한 위치와 역할을 하는 노드들을 설명하기 위해 제시
    * 구조적 공백의 위치에 있는 노드는 다른 노드들과 중복적인 관계를 가지고 있지 않기 때문에 노드와 노드를 연결시켜주는 중요한 역할을 함

    * 실제 조직 내에서 형성된 개인의 인적 네트워크 구조상에서 구조적 공백이 많은 사람이 구조적 공백이 적은 사람의 경우보다 더 빨리 승진한다는 것을 밝힘.
    
* **왓츠, 스트로가츠의 '좁은 세상 네트워크'**
    * 인접한 노드와 일정한 규칙에 따라 일정한 숫자로만 연결되어 있는 정규 네트워크, 무작위로 연결되어 있는 무작위 네트워크의 중간에 존재하는 좁은 세상 네트워크를 제시
    * 노드들 간에 지름길이 존재하며 이것을 통해 전체 네트워크가 긴미라게 연결됨.
    * 무척도 네트워크가 밝혀지게 되는 계기 제공

* **바라바시, 알버트, 정하웅의 '무척도 네트워크'**
    * 하이퍼 링크로 연결된 인터넷의 물리적 연결 구조에 대한 논문에서 인터넷 견결 수 분포가 정규분포를 따르지 않고 멱함수법칙을 따르는 것을 밝힘. 이러한 네트워크 구조를 '무작위 네트워크'와 구별하여 '무척도 네트워크'라고 명명함.
    * 무척도 네트워크에서는 소수의 링크만을 가지고 있는 수많은 노드들과 수많은 링크를 가지고 있는 소수의 노드들로 양분됨.
    * 노드 간에 연결이 이루어질 때, 특정 노드에 대한 선호적 연결현상이 발생하고 반복되게 되면 아무 많은 링크를 가진 소수의 노드들이 형성되게 되고, 이러한 소수의 노드들은 전체 네트워크에서 허브 등의 중요한 역할을 하게 됨.

## 3. 네트워크 이론의 기초

### 3.1. 그래프의 기본 개념
* 네트워크를 표현하는 네트워크 모델링 과정에서 중요하게 사용되는 것이 그래프, 행렬
* 그래프 이론 : 그래프는 노드와 링크로 구성됨
* 무방향 네트워크 : 책, 논문의 공저자, 네트워크, 페이스북의 친구 관계 네트워크
* 방향 네트워크 : 인터넷 주소의 하이퍼링크 네트워크, 전화통화 네트워크, 트위터의 리트윗, 팔로잉 네트워크 등

* 네트워크 그래프에서 노드와 노드 간 관계의 유무만이 아니라 관계의 정도를 함께 표시할 수 있음
* 관계의 유무만 표현 : 이진 네트워크 / 관계의 정도를 함께 표시 : 가중 네트워크

### 3.2. 그래프의 표현 방법
1. 인접행렬
    * 인접한 노드들 간의 연결 관계를 **행렬**로 표현
2. 노드 리스트
    * 인접한 링크가 있는 노들만을 표기하는 방법으로 노드들만의 목록, 무방향
3. 엣지 리스트
    * 노드들을 연결한 링크들의 리스트로 "링크의 목록"
***

    
## 4. 네트워크 유형
* 방향성과 가중치에 따른 분류
* 관계의 위상에 따른 분류(Ring, Mesh, Star, Line 등)
* 네트워크 크기에 따른 분류(노드의 개수, 링크의 개수 등)
* 분석 대상에 따른 분류(전체 vs 하위)

## 5. 소셜 네트워크 분석 지표

### 5.1. 네트워크 기본 속성 분석

### 5.1.1 네트워크 수준의 속성 분석 
* 네트워크 크기 : 노드의 수로 판단함.
    * 연결관계의 수
        * 방향성이 있는 경우: nx(n-1)
        * 방향성이 없는 경우 : nx(n-1)/2
* 네트워크 밀도 : 노드들 간의 연결된 정도
    * 네트워크 밀도 계산
        * 무방향 이진 네트워크의 밀도 : k/{n x (n-1)/2}
        * 방향 이진 네트워크의 밀도 : k/{n x (n-1)}
        * K: 실제 연결된 링크의 수, n: 노드의 수
* 포괄성 : 전체 네트워크의 밀집도를 연결된 노드의 관점에서 표현

### 5.1.2. 노드 수준의 속성 분석
* 연결정도 : 해당 노드에 직접 연결되어 있는 노드들의 개수, 또는 링크의 개수
* 연결강도 : 노드 간 관계(링크)의 강도
* 연결거리 : 두 개의 특정 노드들 사이에 존재하는 다양한 연결경로 중에서 가장 적은 수의 링크 단계를 거치는 경로, 즉 가장 짧은 연결 경로
* 직경 : 네트워크 상의 모든 연결거리 중에서 가장 긴 연결거리
    * 직경이 작을수록 신속한 정보 전달 가능해지고 효율성 높아짐
* 평균연결거리 : 네트워크 내 모든 노드 쌍의 최단 연결거리를 산출하여 평균값을 계산한 것을 해당 네트워크의 평균 연결거리라고 함

### 5.1.3. 네트워크에 내재된 속성 분석
* 상호성
* 군집화 계수


### 5.2. 중심성 분석
* 일반적으로 가장 많이 사용되는 지표로 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표
* 전체 네트워크에서 중요한 역할을 하는 노드가 무엇인지 파악 가능
* 연결정도 중심성 : 한 노드가 얼마나 많은 다른 노드들과 **연결관계**를 맺고 있는지 측정
* 근접 중심성 : 한 노드가 얼마나 전체 네트워크의 **중심에 근접**해 있는지를 측정
* 매개 중심성 : 한 노드가 다른 노드들 간의 네트워크 관계 형성에 있어서 **중개자 혹은 매개자 역할을 얼마나 수행**하는지 측정
* 아이겐 벡터 중심성 : 연결된 다른 노드들의 **중요도**를 함께 반영



```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# 데이터 불러오기
data = pd.read_csv('C:/Users/thgml/Downloads/Pokemon.csv')
# 12개의 열, 행:포켓몬 종류
data.head()
data.tail()

# null 값 처리
print(len(data.isnull().any()))
data.isnull().any()
data['Type 2'].replace(np.nan, '0', inplace = True)
data['Type 2'].head(10)

print("Number of pokemon are: " + str(data['Name'].unique())) #포켓몬 종류 출력
# unique 함수는 고유한 값으로 출력하게 함
name = pd.DataFrame(data['Name'].unique().tolist(), columns = ['Pokemon'])
# data의 Name 을 고유한 값으로 열 이름을 Pokemon으로 지정하여 데이터프레임 만들기
```

```python
npoke_total = data.copy()
print(npoke_total.columns) #컬럼명 확인
npoke_total = pd.concat([npoke_total['Name'], data['Total']], axis = 1)
# 복사한 데이터프레임에서 포켓몬 종류와 종류별로 몇마리인지 확인
npoke_total.head(10)
```

```python
# 종류별 몇 마리인지 시각화
sns.set()
plt.figure(figsize=(8, 20))
ax = sns.barplot(x = 'Total', y = 'Name', data = npoke_total.sort_values(by = 'Total', ascending = False).head(30))
ax.set(xlabel = 'Overall', ylabel = 'Pokemon')
plt.show()
```

```python
# 종류별 네트워크 분석
g = nx.Graph()
g = nx.nx.from_pandas_edgelist(data, source = 'Name', target = 'Type 1')
# type 1(포켓몬 종류 속성)에 대한 네트워크 분석
print(nx.info(g))

plt.figure(figsize = (20, 20))
pos = nx.spring_layout(g, k = 0.15)
nx.draw_networkx(g, pos, with_labels = 'TRUE', node_size = 100, node_color = 'blue')
plt.show()

# 속성이 water인 포켓몬이 가장 많고 flying 이 가장 적음
```
